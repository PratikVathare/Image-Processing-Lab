```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read image in grayscale
img = cv2.imread("/content/drive/MyDrive/Image Processing Lab/nature.jpg", 0)
```

```python
print(img.shape)
```

```python
negative_img = 255 - img
```

```python
r_min = np.min(img)
r_max = np.max(img)

contrast_img = (img - r_min) * (255 / (r_max - r_min))
contrast_img = contrast_img.astype(np.uint8)
```

```python
bit_planes = []

for i in range(8):
    bit_plane = (img >> i) & 1
    bit_planes.append(bit_plane * 255)
```

```python
low = 100
high = 150

gray_slice = img.copy()
gray_slice[(img >= low) & (img <= high)] = 255
```

```python
titles = [
    "Original Image",
    "Negative Image",
    "Contrast Stretched Image",
    "Bit Plane 0 (LSB)",
    "Bit Plane 7 (MSB)",
    "Gray Level Slicing"
]

images = [
    img,
    negative_img,
    contrast_img,
    bit_planes[0],
    bit_planes[7],
    gray_slice
]

plt.figure(figsize=(12, 8))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
