```python
# Task 1: Read an image
image = cv2.imread("/content/drive/MyDrive/Image Processing Lab/eximage.jpg")
if image is None:
    print("Error: Image not loaded")
else:
    print("Image loaded successfully")
```

```python
cv2.imwrite("saved_color_image.png", image)
cv2.imwrite("saved_gray_image.png", gray_image)
```
